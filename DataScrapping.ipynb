```python
import requests
from bs4 import BeautifulSoup
```

```python
dollar_page = requests.get("https://www.xe.com/currencyconverter/convert/?Amount=1&From=USD&To=PKR")
```

```python
soup = BeautifulSoup(dollar_page.content, 'lxml')
dollar_rate = (float)( soup.find("p", {"class": "result__BigRate-sc-1bsijpp-1 iGrAod"}).text.split(" ")[0])
dollar_rate
```

```python
current_datetime = requests.get("https://www.worldtimeserver.com/current_time_in_PK.aspx")
time_soup = BeautifulSoup(current_datetime.content, "lxml")
```

```python
local_time = time_soup.find("div", {"class": "local-time"})
date = local_time.find("h4").text
time = local_time.find("span").text.strip()
datetime = date + "--" + time
print(datetime)
```

```python
query = "apple"
url = "https://www.ebay.com/sch/i.html?_nkw="
ecommerence_page = requests.get(url + query)

soup_page = BeautifulSoup(ecommerence_page.content, 'lxml')
```

```python
products = soup_page.find_all("li", {"class": "s-item s-item__pl-on-bottom"})
len(products)
```

```python
products = products[1:]
products_dollar = []
for product in products:
    price = (float)(product.find("span", {"class": "s-item__price"}).text.split(" ")[0][1:])
    title = product.find("div", {"class": "s-item__title"}).text
    products_dollar.append([title, price, datetime])
    
products_dollar
    
```

```python
import pandas as pd

df_dollar = pd.DataFrame(products_dollar, columns = ["title", "$ in price", "datetime"], index = None)
df_dollar
```

```python
df_dollar.to_excel("dollar_scrap.xlsx", index = False)
```

```python
df = pd.read_excel("dollar_scrap.xlsx")
```
